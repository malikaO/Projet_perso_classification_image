Voici les accuracy en fonction du nombre de cluster tester. 

```python
import pandas as pd
df = pd.DataFrame({'Nombre de cluster': [10,50,100,150,200,250,256,500,600,1100,2000,3000], 'Accuracy': [56.33,79.58, 85.51, 86.31,87.52,89.17,89.11,94.96,92.39,93.52, 94.83, 92.07]}, columns = ['Nombre de cluster','Accuracy'])
```

```python
df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.plot(df['Nombre de cluster'],df['Accuracy'])
plt.xlabel('Nombre de cluster')
plt.ylabel('Accuracy')
plt.title('Accuracy en fonction du nombre de cluster')
plt.show()
```

L'accuracy augmente très lentement en fonction du nombre de cluster. Est ce qu-il serait judicieux de choisir un plus grand nombre de cluster pour augmenter de peu mon accuracy où dois-je changer de stratégie?
